**Début de l'implémentation de l'algorithme de clustering K-Means from scratch**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("iris.csv")
```

```python
df
```

Etant donné que le K-Means est un algorithme d'apprentissage non supervisé, on enlève donc la target comportant les noms de fleurs.

```python
df = df.drop(['species'], axis = 1)
```

```python
df
```

On standardise nos données également

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
X = df[['sepal_length','sepal_width','petal_length','petal_width']].values
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
def distance_euclidienne(x, y):
    distance = 0
    for k in range(len(x)):
        distance += (x[k]-y[k])**2
    return np.sqrt(distance)
        
```

```python
class k_means:

    def __init__(self, k=3, max_iter=100):
        self.k = k
        self.max_iter = max_iter
        self.centroids = None
        self.inerties = []

        
    def inertie(self, X, centroids):
        inertie = 0
        for x in X:
            distances = [distance_euclidienne(x, c) for c in centroids]
            nearest = np.argmin(distances)
            inertie += distances[nearest]**2
        return inertie
        
    def fit(self, X):

        centroids = np.random.uniform(
            low=X.min(axis=0),
            high=X.max(axis=0),
            size=(self.k, X.shape[1])
        )

        for w in range(self.max_iter):

            old_centroids = centroids.copy()

            
            clusters = [[] for _ in range(self.k)]

            
            for x in X:

                distances = [
                    distance_euclidienne(x, centroids[i])
                    for i in range(len(centroids))
                ]

                cluster_index = np.argmin(distances)
                clusters[cluster_index].append(x)

        
            new_centroids = []

            for cluster in clusters:

                if len(cluster) > 0:
                    new_centroids.append(np.mean(cluster, axis=0))
                else:
                    new_centroids.append(
                        X[np.random.randint(len(X))]
                    )

            centroids = np.array(new_centroids)
            
            self.inerties.append(self.inertie(X, centroids)) 

            if np.allclose(old_centroids, centroids):
                break

        self.centroids = centroids
        return self
    
    
    def predict(self, X):

        predictions = []

        for x in X:

            distances = [
                distance_euclidienne(x, c)
                for c in self.centroids
            ]

            predictions.append(np.argmin(distances))

        return predictions
```

```python
model = k_means(k=3).fit(X)


print(model.inerties)


import matplotlib.pyplot as plt

plt.plot(model.inerties)
plt.xlabel("Iterations")
plt.ylabel("Inertie")
plt.title("Évolution de l'inertie")
plt.show()
```

```python
labels = model.predict(X)

plt.scatter(
    X[:,0],
    X[:,1],
    c=labels
)

plt.scatter(
    model.centroids[:,0],
    model.centroids[:,1],
    c='red',
    marker='x',
    s=200
)

plt.show()
```
